```r
set.seed(123)
train_idx <- sample(1:nrow(Boston), 0.8 * nrow(Boston))
train <- Boston[train_idx, ]
test <- Boston[-train_idx, ]
```

```r
# Load MLflow
library(mlflow)

# Model 1: Linear Regression
model_lm <- lm(medv ~ ., data = train)

# Model 2: Decision Tree
library(rpart)
model_tree <- rpart(medv ~ ., data = train)

# Model 3: Random Forest
library(randomForest)
model_rf <- randomForest(medv ~ ., data = train)
```

```r
## Linear Regression Model LM 
summary(model_lm)
```

```r
# Decision Tree model summary
summary(model_tree)
```

```r
summary(model_rf)
```

```r
## Linear Regression predictions model 
predictions_lm <- predict(model_lm, newdata = test)

# Check first few predictions
head(predictions_lm)
```

```r
## Decision Tree predictions model
predictions_tree <- predict(model_tree, newdata = test)
head(predictions_tree)
```

```r
predictions_rf <- predict(model_rf, newdata = test)
head(predictions_tree)
```

```r
## Raw Evalutation 
### Evaluation for Linear Regression 
mse_lm <- mean((predictions_lm - test$medv)^2)
mse_lm

# Calculate R-squared on test set
ss_total_lm <- sum((test$medv - mean(test$medv))^2)
ss_res_lm <- sum((predictions_lm - test$medv)^2)
r_squared_lm <- 1 - (ss_res_lm/ss_total_lm)
r_squared_lm
```

```r
## Raw Evalutation 
### Evaluation for Decision Tree Regression 
mse_tree <- mean((predictions_tree - test$medv)^2)
mse_tree

# Calculate R-squared on test set
ss_total_tree <- sum((test$medv - mean(test$medv))^2)
ss_res_tree <- sum((predictions_tree - test$medv)^2)
r_squared_tree <- 1 - (ss_res_tree/ss_total_tree)
r_squared_tree
```

```r
## Raw Evalutation 
### Evaluation for Random Forest  
mse_rf <- mean((predictions_rf - test$medv)^2)
mse_rf

# Calculate R-squared on test set
ss_total_rf <- sum((test$medv - mean(test$medv))^2)
ss_res_rf <- sum((predictions_rf - test$medv)^2)
r_squared_rf <- 1 - (ss_res_rf/ss_total_rf)
r_squared_rf
```

```r
library(mlflow)
library(carrier)
library(randomForest)
library(MASS)  # for Boston dataset
Sys.setenv(
  AWS_ACCESS_KEY_ID = "root",
  AWS_SECRET_ACCESS_KEY = "12345678",
  MLFLOW_S3_ENDPOINT_URL = "http://127.0.0.1:9000"
)

set.seed(123)

# Data
data(Boston)
train_idx <- sample(1:nrow(Boston), 0.8 * nrow(Boston))
train <- Boston[train_idx, ]
test  <- Boston[-train_idx, ]

# Train RandomForest
model_rf <- randomForest(medv ~ ., data = train)

# Evaluate
pred_rf <- predict(model_rf, newdata = test)
mse_rf <- mean((pred_rf - test$medv)^2)

# ---- Log with MLflow ----
with(mlflow_start_run(), {
  mlflow_log_param("model_type", "random_forest")
  mlflow_log_metric("mse", mse_rf)

  # ✅ Crate with explicit namespace
rf_crate <- carrier::crate(
  function(df) {
    predict(model_rf, df)   # model_rf is captured in environment
  },
  model_rf = model_rf
)
  # Save locally + log to MLflow
  mlflow_save_model(rf_crate, path = "rf_model")
  mlflow_log_model(rf_crate, artifact_path = "model")
})
print(done)
```
